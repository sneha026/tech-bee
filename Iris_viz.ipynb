```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_iris= pd.read_csv("IRIS.csv")
```

```python
df_iris.info()
```

```python
df_iris.isnull().sum()
```

```python
df_iris.nunique()
```

```python
##Learn the visualization
df_iris.plot(kind='scatter', x= 'sepal_length',y='sepal_width')
```

```python
##sns plot

sns.jointplot(x='sepal_length',y='sepal_width',data=df_iris,size=5)
```

```python
##FacetGrid

sns.FacetGrid(df_iris, hue="species",palette='husl',size=5)\
   .map(plt.scatter,"sepal_length","sepal_width")\
    .add_legend()
    
```

```python
##Boxplot
sns.boxplot(x="species",y="petal_length",palette="husl",data=df_iris)
```

```python
##violinplot
sns.violinplot(x="species",y="petal_length",palette="husl",data=df_iris)
```

```python
##swarm plot over violin plot
sns.violinplot(x="species",y="petal_length",palette="husl",data=df_iris)
sns.swarmplot(x="species",y="petal_length",data=df_iris,color='white',alpha=.9)
```

```python
sns.FacetGrid(df_iris,hue="species",palette="husl",size=6).map(sns.kdeplot,'petal_length').add_legend()
```

```python
##pairplot

sns.pairplot(hue="species",palette="husl",size=5,data=df_iris)
```

```python
df_iris.boxplot(by='species',figsize=(10,12))
```

```python
from pandas.plotting import andrews_curves
andrews_curves(df_iris, "species")
```

```python
##Multivariate vizualization techniques
# Parallel coordinates plots each feature on a separate column & then draws lines
# connecting the features for each data sample

from pandas.plotting import parallel_coordinates
parallel_coordinates(df_iris, "species")
```

```python
 #A final multivariate visualization technique pandas has is radviz
# Which puts each feature as a point on a 2D plane, and then simulates
# having each sample attached to those points through a spring weighted
# by the relative value for that feature

from pandas.plotting import radviz
radviz(df_iris, "species")
```
